# 04 — Embodied Reasoning: Robot Next-Action Prediction

**What this notebook demonstrates:**
- Image-based reasoning with `<image>` tags
- Robot action prediction ("What should the robot do next?")
- Physical world understanding for embodied AI

**Key concept:** Cosmos-Reason2 is trained on physical-world data including
robot manipulation, navigation, and human activity. It can predict next actions
given a camera observation — useful for planning in robotics.

---

```python
import os
os.environ['QWEN_VL_VIDEO_READER'] = 'torchcodec'

import sys, os, time
sys.path.insert(0, os.path.join(os.getcwd(), '..'))

import torch
print(f"🎮 GPU: {torch.cuda.get_device_name(0)}")
```

```python
project_root = os.path.abspath('..')
sample_image = os.path.join(project_root, 'sample.png')
print(f"🖼  Image: {sample_image}")
print(f"   Exists: {os.path.exists(sample_image)}")
print(f"   Size: {os.path.getsize(sample_image) / 1e6:.2f} MB")
```

## Display the Image

Let's see what the "robot" is looking at:

```python
from PIL import Image
from IPython.display import display

img = Image.open(sample_image)
print(f"   Resolution: {img.size[0]}x{img.size[1]}")
# Display at reduced size for notebook
display(img.resize((640, int(640 * img.size[1] / img.size[0]))))
```

## Load Vision Model with Reasoning

```python
from strands import Agent
from strands_cosmos import CosmosVisionModel

t0 = time.time()
model = CosmosVisionModel(
    model_id="nvidia/Cosmos-Reason2-2B",
    reasoning=True,
    params={"max_tokens": 2048, "temperature": 0.6},
)
agent = Agent(model=model)
print(f"✅ Ready in {time.time() - t0:.1f}s")
```

## Next-Action Prediction

The canonical Cosmos-Reason2 embodied prompt: *"What can be the next immediate action?"*

```python
t0 = time.time()
result = agent(f"<image>{sample_image}</image> What can be the next immediate action?")
print(f"\n⏱  Time: {time.time() - t0:.1f}s")
```

## Scene Understanding

```python
t0 = time.time()
result = agent(
    f"<image>{sample_image}</image> "
    "Describe all objects in the scene, their spatial relationships, "
    "and what a robot arm could manipulate."
)
print(f"\n⏱  Time: {time.time() - t0:.1f}s")
```
