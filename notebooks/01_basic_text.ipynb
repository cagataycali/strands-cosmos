## Load the Model

`CosmosModel` is the **text-only** provider. It wraps NVIDIA Cosmos-Reason2 
(based on Qwen3-VL architecture) and implements the `strands.models.Model` interface.

On first run, the model downloads from HuggingFace (~5GB for 2B).

```python
from strands import Agent
from strands_cosmos import CosmosModel

t0 = time.time()
model = CosmosModel(model_id="nvidia/Cosmos-Reason2-2B")
print(f"✅ Model loaded in {time.time() - t0:.1f}s")
```

## Create an Agent

The Strands `Agent` wraps any model provider. Once created, you just call it with natural language.

```python
agent = Agent(model=model)
print("✅ Agent ready")
```

## Physics Reasoning

Cosmos-Reason2 is trained for physical world understanding. Let's test basic physics reasoning.

```python
t0 = time.time()
result = agent("Explain the physics of a ball rolling down a ramp. Be concise.")
print(f"\n⏱  Generation time: {time.time() - t0:.1f}s")
```

## Try Another Prompt — Gravity & Free Fall

```python
t0 = time.time()
result = agent("What happens when you drop a feather and a hammer on the Moon? Why?")
print(f"\n⏱  Generation time: {time.time() - t0:.1f}s")
```
